```python
from sage.repl.display.formatter import SagePlainTextFormatter
f = SagePlainTextFormatter()

#function to check whether TDAG or not 
#input has to be the adjacency matrix of a DAG
def is_TDAG(A):
    m = A.nrows()
    B = (Matrix.identity(m) - A).inverse()
    for i in range(0,m):
        for j in range(0,m):
            if (i == j):
                continue
            if (B[i][j]*A[i][j] == 0):
                if(B[i][j]+A[i][j] == 0):
                    return True
                else:
                    return False
            else:
                return True

#function to remove the diagonal 1's 
#will give the adjacency matrix
def trim(M):
    m = M.nrows()
    return((M - Matrix.identity(m)).transpose())

#function to get DAG from matrix as described in PDF
#this is the map \mathfrak{G}
def get_graph(M):
    return DiGraph(trim(M))

#function to get matrix from DAG as described in PDF
#this is the map \mathrfak{M}
def get_matrix(G):
    A = G.adjacency_matrix()
    A = A + Matrix.identity(A.nrows())
    return(A.transpose())

#function to find the tensor of two DAGs
def tensor_DAG(G1, G2):
    A1 = get_matrix(G1)
    A2 = get_matrix(G2)
    return get_graph(A1.tensor_product(A2))

def no_edge(n):
    return DiGraph([[i for i in range(1,n+1)],[]])

def dist(X,Y):
    return (X-Y).norm('frob')
    
M1 = Matrix([[1,0,1],[0,1,1],[0,0,1]])
M2 = Matrix([[1,1],[0,1]]) 
H1 = DiGraph([(3,2),(3,1)])
```
