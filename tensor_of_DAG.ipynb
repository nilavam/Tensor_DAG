```python
#function to remove the diagonal 1's - will give the adjacency matrix
def trim(M):
    m = M.nrows()
    return((M - Matrix.identity(m)).transpose())

#function to get DAG from matrix as described in PDF - this is the map \mathrfak{G}
def get_graph(M):
    return DiGraph(trim(M))

#function to get matrix from DAG as described in PDF - this is the map \mathrfak{M}
def get_matrix(G):
    A = G.adjacency_matrix()
    A = A + Matrix.identity(A.nrows())
    return(A.transpose())

#function to find the tensor of two DAGs
def tensor_DAG(G1, G2):
    A1 = get_matrix(G1)
    A2 = get_matrix(G2)
    return get_graph(A1.tensor_product(A2))
    
M1 = Matrix([[1,0,1],[0,1,1],[0,0,1]])
M2 = Matrix([[1,1],[0,1]]) 
G1 = DiGraph([(3,1),(3,2)])
G2 = DiGraph([(2,1)])
tensor_DAG(G1,G1)
```
